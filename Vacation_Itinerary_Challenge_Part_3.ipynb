```python
#file to load
weather_data_to_load ="data/WeatherPy_vacation.csv"
```

```python
#read csv files into a df
weather_data_df=pd.read_csv(weather_data_to_load)
weather_data_df.head(10)
```

```python
#Cities selected for directions: Mahebourg,Souillac,Bambous Virieux, Grand Gaube
#syntax df.loc[df[‘column name’] condition]
#df for Mahebourg
weather_data_df_c1=weather_data_df.loc[weather_data_df['City']=='Mahebourg']
weather_data_df_c1
```

```python
#df for Souillac
weather_data_df_c2=weather_data_df.loc[weather_data_df['City']=='Souillac']
weather_data_df_c2
```

```python
#df for Bambous Virieux
weather_data_df_c3=weather_data_df.loc[weather_data_df['City']=='Bambous Virieux']
weather_data_df_c3
```

```python
#df for Grand Gaube
weather_data_df_c4=weather_data_df.loc[weather_data_df['City']=='Grand Gaube']
weather_data_df_c4
```

```python
#pd.concat([s1, s2])
shortlist_df=pd.concat([weather_data_df_c1,weather_data_df_c2,weather_data_df_c3,weather_data_df_c4])
shortlist_df
```

```python
# Import the API key.
from config import g_key
gmaps.configure(api_key=g_key)

#lat and lng from weather_data_df_c1,weather_data_df_c2,weather_data_df_c3,weather_data_df_c4
lat_1=weather_data_df_c1.loc[weather_data_df_c1.City=="Mahebourg",'Lat'].values[0]
lng_1=weather_data_df_c1.loc[weather_data_df_c1.City=="Mahebourg",'Lng'].values[0]
lat_2=weather_data_df_c2.loc[weather_data_df_c2.City=="Souillac",'Lat'].values[0]
lng_2=weather_data_df_c2.loc[weather_data_df_c2.City=="Souillac",'Lng'].values[0]
lat_3=weather_data_df_c3.loc[weather_data_df_c3.City=="Bambous Virieux",'Lat'].values[0]
lng_3=weather_data_df_c3.loc[weather_data_df_c3.City=="Bambous Virieux",'Lng'].values[0]
lat_4=weather_data_df_c4.loc[weather_data_df_c4.City=="Grand Gaube",'Lat'].values[0]
lng_4=weather_data_df_c4.loc[weather_data_df_c4.City=="Grand Gaube",'Lng'].values[0]

# Latitude-longitude pairs
mahebourg=(lat_1,lng_1)
souillac=(lat_2,lng_2)
bambos=(lat_3,lng_3)
grand_gaube=(lat_4,lng_4)

#create dataframe hotel_df to store hotel names along with city, country, max temp, and coordinates
hotel_df = shortlist_df[["City", "Current description","Country", "Max Temp", "Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""


# Set the parameters to search for a hotel in Paris.
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key}

#iterate through dataframe
for index, row in hotel_df.iterrows():
    #get latitude and longitude
    lat = row["Lat"]
    lng = row["Lng"]

    #add the latitude and longitude to location key for the params dictionary in gmaps
    params["location"] = f"{lat},{lng}"

    
     # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except(IndexError):
        print("Hotel not found... skipping.")

hotel_df
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current description</dt><dd>{Current description} °F</dd>
</dl>
"""
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
#syntax reference
#fig = gmaps.figure()
#geneva2zurich_via_montreux = gmaps.directions_layer(geneva, zurich, waypoints=[montreux],travel_mode='BICYCLING')
#fig.add_layer(geneva2zurich_via_montreux)
#fig

fig = gmaps.figure()
directions_four = gmaps.directions_layer(mahebourg,souillac,waypoints=[bambos,grand_gaube],travel_mode="DRIVING")
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(directions_four)
fig.add_layer(marker_layer)
fig
```
